```python
import pandas as pd
import ast
import json
# 读取文件
file_path = '../data/output_data/output1.xlsx'
df = pd.read_excel(file_path)



# 创建一个空的列表来存储过滤后的数据
filtered_data = []

# 遍历每一行
for i in range(len(df)):
    row = df.iloc[i]
    
    # 只处理 label=0 的行
    if row['label'] == 0:
        gpt4o_result = row['gpt4o_result']
        
        # 检查 gpt4o_result 是否为字符串类型
        if isinstance(gpt4o_result, str):
            try:
                # 尝试解析 gpt4o_result 列
                result_list = json.loads(gpt4o_result)
            except json.JSONDecodeError:
                # 解析失败，继续处理下一行
                continue
        else:
            # 如果不是字符串类型，跳过当前行
            continue
        
        # 如果 result_list 是列表
        if isinstance(result_list, list):
            # 如果只有一个字典并且 type 既不是“发起”也不是“讲解”
            if len(result_list) == 1 and result_list[0]['type'] not in ['发起', '讲解']:
                # 删除当前行和紧接着的 label=1 的行（学生回应）
                if i + 1 < len(df) and df.iloc[i + 1]['label'] == 1:
                    continue  # 跳过下一个 label=1 的行
                continue  # 跳过当前行
            else:
                # 如果符合条件，保留当前行数据
                filtered_data.append(row)
    else:
        # 如果是 label=1 的行，保留当前行数据
        filtered_data.append(row)

# 将过滤后的数据转换为 DataFrame
filtered_df = pd.DataFrame(filtered_data)
filtered_df
# # 保存过滤后的数据
# filtered_output_path = '/mnt/data/filtered_output.xlsx'
# filtered_df.to_excel(filtered_output_path, index=False)
# 
# filtered_output_path
```

```python
def parse_gpt4o_result(gpt4o_result):
    if pd.isnull(gpt4o_result):
        return None
    try:
        # 去除多余的换行符和转义字符
        s = gpt4o_result.replace('\n', '').replace('\\n', '').replace('\\\\', '\\')
        # 解析JSON字符串
        return json.loads(s)
    except json.JSONDecodeError as e:
        print(f"JSON解析错误: {e}")
        print(f"原始字符串: {gpt4o_result}")
        return None

# 应用解析函数
df['parsed_gpt4o_result'] = df['gpt4o_result'].apply(parse_gpt4o_result)


```

```python
def filter_and_merge(parsed_gpt4o_result):
    if parsed_gpt4o_result is None:
        return None, False
    result = parsed_gpt4o_result.get('result', [])
    # 过滤出 type 为 "发起" 或 "讲解" 的字典
    filtered_result = [d for d in result if d.get('type') in ['发起', '讲解']]
    # 判断是否需要删除行
    if len(result) == 1 and result[0].get('type') not in ['发起', '讲解']:
        # 标记该行用于删除
        return None, True  # 第二个值表示需要删除
    if not filtered_result:
        # 没有 "发起" 或 "讲解" 的字典
        return None, False
    # 合并相邻相同类型的字典
    merged_result = []
    prev_type = None
    merged_content = ''
    for d in filtered_result:
        curr_type = d.get('type')
        content = d.get('content', '')
        if curr_type == prev_type:
            merged_content += content
        else:
            if prev_type is not None:
                merged_result.append({'type': prev_type, 'content': merged_content})
            prev_type = curr_type
            merged_content = content
    # 添加最后一个合并后的字典
    if prev_type is not None:
        merged_result.append({'type': prev_type, 'content': merged_content})
    # 返回新的 filtered_gpt4o_result
    return {'result': merged_result, 'compliance': parsed_gpt4o_result.get('compliance')}, False

# 应用过滤和合并函数
filtered_results = []
delete_flags = []

for idx, row in df.iterrows():
    parsed_gpt4o_result = row['parsed_gpt4o_result']
    filtered_gpt4o_result, delete_flag = filter_and_merge(parsed_gpt4o_result)
    filtered_results.append(filtered_gpt4o_result)
    delete_flags.append(delete_flag)

df['filtered_gpt4o_result'] = filtered_results
df['delete_flag'] = delete_flags
```

```python
# 第四步：删除指定的行
# 初始化一个列用于标记需要删除的行
df['to_delete'] = False

for idx in df.index:
    if df.at[idx, 'delete_flag']:
        df.at[idx, 'to_delete'] = True
        # 查找下一个 label=1 的行
        next_idx = idx + 1
        if next_idx in df.index and df.at[next_idx, 'label'] == 1:
            df.at[next_idx, 'to_delete'] = True

# 删除标记为需要删除的行
df = df[df['to_delete'] == False].reset_index(drop=True)
```
